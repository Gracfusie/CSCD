```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
```

```python
file_path = 'rdata_output.txt'
with open(file_path, 'r') as f:
    data = f.read().splitlines()
result = []
for i in range(len(data)):
    data_1 = data[i][0:8]
    data_2 = data[i][8:16]
    data_3 = data[i][16:24]
    data_4 = data[i][24:32]
    result.append(int(data_1, 2))
    result.append(int(data_2, 2))
    if (i % 3 != 2):
        result.append(int(data_3, 2))
        result.append(int(data_4, 2))
result = np.array(result)
result = result.reshape(13, 14, 10)
# 转置
result = result.transpose(2, 1, 0)
result = torch.from_numpy(result)
```
